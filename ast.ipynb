### Since the salaries are in ranges and also strings, here is the code to extract salaries in numbers!

```python
import pandas as pd
import numpy as np
import re

def calculate_mean_and_add_column(file_path, column_name):
    try:
        df = pd.read_excel(file_path)
    except Exception as e:
        print("Error:", e)
        return

    def calculate_mean(row):
        try:
            numbers = re.findall(r'₹([\d,]+)', row[column_name])
            if not numbers:
                return None
            numbers_int = [int(number.replace(',', '')) for number in numbers]
            mean = np.mean(numbers_int)
            return mean
        except Exception as e:
            return None

    df['Approx salary'] = df.apply(calculate_mean, axis=1)
    
    df.to_excel(file_path, index=False)

    print("Approx salary column added and changes saved to the same Excel file.")

# Example usage
file_path = "C:/Users/gowth/Downloads/Output.xlsx"
column_name = 'Salary Range'

calculate_mean_and_add_column(file_path, column_name)
```
